```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
titles=soup.find_all(['h1','h2','h3','h4','h5','h6'])
print('List all the header tags:',*titles,sep='\n\n')
```
